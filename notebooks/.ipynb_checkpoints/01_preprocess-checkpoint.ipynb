```python
with gzip.open(
    "../data/raw/GSE256068_series_matrix.txt.gz",
    "rt"
) as f:
    
    lines = f.readlines()

title_line = [x for x in lines if x.startswith("!Sample_title")][0]
geo_line = [x for x in lines if x.startswith("!Sample_geo_accession")][0]
```

```python
titles = title_line.strip().split("\t")[1:]
gsm_ids = geo_line.strip().split("\t")[1:]
```

```python
titles[:5]
```

```python
gsm_ids[:5]
```

```python
metadata = pd.DataFrame({
    "sample_id": gsm_ids,
    "sample_title": titles
})

metadata["sample_id"] = metadata["sample_id"].str.replace('"', '', regex=False)
metadata["sample_title"] = metadata["sample_title"].str.replace('"', '', regex=False)

metadata.head()
```

```python
metadata["group"] = metadata["sample_title"].str.split("_").str[0]

metadata["binary_class"] = metadata["group"].apply(
    lambda x: "Control" if "Control" in x else "Epilepsy"
)

metadata.head()
```

```python
metadata["group"].value_counts()
metadata["binary_class"].value_counts()
```

```python
metadata.to_csv("../data/processed/metadata.csv", index=False)
```
